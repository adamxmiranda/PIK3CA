```r
PIK3CA_E545K_IDs <- meta_mutations@data %>%
              filter(Hugo_Symbol == "PIK3CA") %>%
              filter(HGVSp_Short == "p.E545K") %>%
              select(Tumor_Sample_Barcode)
PIK3CA_E545K_IDs <- levels(PIK3CA_E545K_IDs$Tumor_Sample_Barcode)[as.numeric(PIK3CA_E545K_IDs$Tumor_Sample_Barcode)]
PIK3CA_E545K_IDs

PIK3CA_H1047R_IDs <- meta_mutations@data %>%
              filter(Hugo_Symbol == "PIK3CA") %>%
              filter(HGVSp_Short == "p.H1047R") %>%
              select(Tumor_Sample_Barcode)
PIK3CA_H1047R_IDs <- levels(PIK3CA_H1047R_IDs$Tumor_Sample_Barcode)[as.numeric(PIK3CA_H1047R_IDs$Tumor_Sample_Barcode)]
PIK3CA_H1047R_IDs
PIK3CA_H1047R_IDs
```
