```r
AREG_eQTL <- read_csv(file = "/data/park_lab/adam/genomeTrackPlot/GTex_eQTL_AREG.csv") %>%
             as.data.frame()
AREG_eQTL
```

```r
chrom <-  str_split_i(string = AREG_eQTL$`Variant Id`, pattern = "_", i = 1)
start <-  str_split_i(string = AREG_eQTL$`Variant Id`, pattern = "_", i = 2) %>%
          as.numeric() -100
end <-  str_split_i(string = AREG_eQTL$`Variant Id`, pattern = "_", i = 2) %>%
          as.numeric() +100
eQTLbars <- cbind(chrom, start, end, AREG_eQTL$`SNP Id`) %>% as.data.frame()
names(eQTLbars) <- c("chrom", "start", "end", "peak_ID")
eQTLbars$start <- as.numeric(eQTLbars$start)
eQTLbars$end <- as.numeric(eQTLbars$end)
eQTLbars
```
