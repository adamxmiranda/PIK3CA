## High Virus R Analysis

```r
setwd(dir = "/data/park_lab/adam/CRISPRscreen/trimmed_reads")
library(tidyverse)
```

```r
library(tidyverse)
```

```r
Hi_virus <- read_tsv(file = "High_Virus.gene_summary.txt")
Hi_virus
```

```r
Hi_virus <- as.data.frame(Hi_virus)
```

```r
Hi_virus$Gene
```

```r
library(ggrepel)
```

```r
Hi_virus_plot <- ggplot(data = Hi_virus, aes(x = `Ex9_Hi|beta`, y = `Ex20_Hi|beta`, label = Gene)) +
            geom_point() + geom_abline(slope = 1) + geom_label_repel()
Hi_virus_plot
```

```r
Hi_virus$betadiff <- abs(Hi_virus$`Ex9_Hi|beta` - Hi_virus$`Ex20_Hi|beta`)
```

```r
Hi_virus
```

```r
betadiff_hist <- ggplot(Hi_virus, aes(x = betadiff)) +
                 geom_histogram(binwidth = 0.005)
betadiff_hist
```

```r
points2label <- Hi_virus %>%
                dplyr::filter(betadiff > 0.12)
points2label
```

```r
library(wesanderson)
```

```r
Hi_virus_plot <- ggplot(data = Hi_virus, aes(x = `Ex9_Hi|beta`, y = `Ex20_Hi|beta`, 
                                             label = Gene)) +
                 theme_bw() + theme(legend.position = "none") +
                 geom_point(aes(colour = Gene %in% points2label$Gene)) + 
                 scale_colour_manual(values = wes_palette("Royal1")) + 
                 geom_abline(slope = 1) + 
                 geom_abline(slope = 1, intercept = 0.12, colour = "red") + 
                 geom_abline(slope = 1, intercept = -0.12, colour = "red") +
                 geom_label_repel(data = points2label) + 
                 ggtitle(label = "High Viral Transduction", 
                         subtitle = "WT 30uL, Ex9 30uL, Ex20 120uL") +
                 xlab(label = "E545K Beta Score") + ylab(label = "H1047R Beta Score")
Hi_virus_plot
```

```r
pdf(file = "Hi_virus_plot.pdf", width = 8, height = 8)
Hi_virus_plot
dev.off()
```

```r
Hi_virus$Zdiff <- abs(Hi_virus$`Ex9_Hi|z` - Hi_virus$`Ex20_Hi|z`)
Lo_virus$Zdiff <- abs(Lo_virus$`Ex9_Lo|z` - Lo_virus$`Ex20_Lo|z`)
Total_virus$Zdiff <- abs(Total_virus$`Ex9|z` - Total_virus$`Ex20|z`)
```

```r
Zdiff_hist_Hi <- ggplot(Hi_virus, aes(x = Zdiff)) +
                 geom_histogram(binwidth = 0.005) + ggtitle(label = "High Transduction Z-diff")
Zdiff_hist_Hi
Zdiff_hist_Lo <- ggplot(Lo_virus, aes(x = Zdiff)) +
                 geom_histogram(binwidth = 0.005) + ggtitle(label = "Low Transduction Z-diff")
Zdiff_hist_Lo
Zdiff_hist_Total <- ggplot(Total_virus, aes(x = Zdiff)) +
                 geom_histogram(binwidth = 0.005) + ggtitle(label = "Total Z-diff")
Zdiff_hist_Total
```

```r
Total_virus$`Ex9|z`
```
