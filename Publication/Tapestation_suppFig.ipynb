```r
library(tidyverse)
library(bioanalyzeR)
library(XML)
library(base64enc)
```

```r
setwd("/data/park_lab/adam/Tapestation_suppfig/")
list.files()
```

```r
EnhDeletion_D5000 <- read.tapestation(xml.file = "2023-02-28 - 15-16-18-D5000.xml",
                    csv.file = "2023-02-28 - 15-16-18-D5000_Electropherogram.csv")
qplot.electrophoresis(EnhDeletion_D5000,
    y = "concentration",
    include.markers = TRUE,
    show.peaks = "markers")
```

```r
EnhDeletion_D5000$samples
```

```r
EnhDeletion_D5000$samples$Media <- c("Ladder", rep("20ng/mL EGF", 15), "Ladder", 
                                     rep("20ng/mL EGF", 3), 
                                     rep("EGF-free Media", 12), "Ladder",
                                     rep("EGF-free Media", 6))
EnhDeletion_D5000$samples$Treatment <- c("Ladder",
                                         rep("IDT Negative\nControl #1", 9),
                                         rep("AREG\nEnhancer Flanking", 6),
                                         "Ladder",
                                         rep("AREG\nEnhancer Flanking", 3),
                                         rep("IDT Negative\nControl #1", 9),
                                         rep("AREG\nEnhancer Flanking", 3),
                                         "Ladder",
                                         rep("AREG\nEnhancer Flanking", 6))
EnhDeletion_D5000$samples$Genotype <- c("Ladder",
                                        "WT","E545K", "H1047R", "WT", "E545K", "H1047R",
                                        "WT","E545K", "H1047R", "WT", "E545K", "H1047R",
                                        "WT", "E545K", "H1047R", 
                                        "Ladder",
                                        "WT", "E545K", "H1047R",
                                        "WT", "E545K", "H1047R", "WT", "E545K", "H1047R",
                                        "WT", "E545K", "H1047R", "WT", "E545K", "H1047R",
                                        "Ladder",
                                        "WT", "E545K", "H1047R", "WT", "E545K", "H1047R"
                                        )
EnhDeletion_D5000$samples
```

```r
p <- qplot.electrophoresis(EnhDeletion_D5000,
    y = "concentration", 
    xlim = c(50,1000),
    facets = ~Genotype + Media + Treatment,
    include.markers = TRUE,
    show.peaks = "markers")
```

```r
p2 <- p + coord_cartesian(ylim = c(0,0.3)) +
    theme_classic() + theme(text = element_text(family = "Arial"),
                          axis.title = element_text(size = 24),
                          axis.text = element_text(size = 14, colour = "black"),
                          axis.text.x = element_text(angle = 60, hjust = 1),
                          axis.text.y = element_text(vjust = 0.1),
                          plot.title = element_text(size = 24),
                          plot.subtitle = element_text(size = 16, colour = "black"),
                          strip.text = element_text(size = 12),
                          legend.text = element_text(size = 14),
                          legend.title = element_text(size = 16),
                          legend.position = "none",
                          plot.margin = margin(10, 10, 10, 10))
```

```r
ggsave(filename = "/data/park_lab/adam/figures/tapestation_supp6-21.pdf", plot = p2,
       device = cairo_pdf, height = 8, width = 8)
```
