# Load Libraries and data

```r
library(tidyverse)
```

```r
setwd("/data/park_lab/adam/SyntheticLethal/")
```

```r
data_2_23 <- read_csv("adamdata_Summary__Sample Set (2-23-2024)_20240223_161038.csv")
```

```r
data_2_23
```

# Data Cleaning

```r
data_2_23select <- data_2_23 %>% select(c(1,2))
data_2_23select$Genotype <- c(rep("10AWT", 12),
                              rep("10AE545K", 12),
                              rep("DLD1", 12),
                              rep("MCF7Par", 12),
                              rep("MCF7Cor", 12))
data_2_23select$treatment <- c(rep(
                         c(rep("Negative", 6), rep("Enhancer", 6)),
                         5))
data_2_23select
```

# MCF7 Parental

```r
MCF7parRes <- data_2_23select %>% filter(Genotype == "MCF7Par")
MCF7parResNeg <- MCF7parRes %>% filter(treatment == "Negative")
MCF7parNormFact <- mean(MCF7parResNeg$`Cell count`)
MCF7parRes <- MCF7parRes %>% mutate(Norm = `Cell count`/MCF7parNormFact)
MCF7parRes
```

# MCF7 Corrected

```r
MCF7corRes <- data_2_23select %>% filter(Genotype == "MCF7Cor")
MCF7corResNeg <- MCF7corRes %>% filter(treatment == "Negative")
MCF7corNormFact <- mean(MCF7corResNeg$`Cell count`)
MCF7corRes <- MCF7corRes %>% mutate(Norm = `Cell count`/MCF7corNormFact)
MCF7corRes
```

# DLD-1

```r
DLD1Res <- data_2_23select %>% filter(Genotype == "DLD1")
DLD1ResNeg <- DLD1Res %>% filter(treatment == "Negative")
DLD1NormFact <- mean(DLD1ResNeg$`Cell count`)
DLD1Res <- DLD1Res %>% mutate(Norm = `Cell count`/DLD1NormFact)
DLD1Res
```
