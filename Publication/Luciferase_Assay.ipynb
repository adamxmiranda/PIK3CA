```r
setwd("/data/park_lab/adam/LuciferaseAssay/")
library(tidyverse)
```

```r
results_df <- read_tsv("LuciferaseAssayResults9-7.txt") %>% as.data.frame()
```

```r
Firefly_results_df <- results_df %>% filter(Reporter == "Firefly")
Firefly_results_df
```

```r
WTFirefly_results_df <- Firefly_results_df %>% filter(Genotype == "WT")
E545KFirefly_results_df <- Firefly_results_df %>% filter(Genotype == "E545K")
H1047RFirefly_results_df <- Firefly_results_df %>% filter(Genotype == "H1047R")
WTFirefly_results_df_NULL <- WTFirefly_results_df %>% filter(Plasmid == "Null")
E545KFirefly_results_df_NULL <- E545KFirefly_results_df %>% filter(Plasmid == "Null")
H1047RFirefly_results_df_NULL <- H1047RFirefly_results_df %>% filter(Plasmid == "Null")
WT_Nullav <- mean(WTFirefly_results_df_NULL$Luminescence)
E545K_Nullav <- mean(E545KFirefly_results_df_NULL$Luminescence)
H1047R_Nullav <- mean(H1047RFirefly_results_df_NULL$Luminescence)
```

```r
WTFirefly_results_df <- WTFirefly_results_df %>% mutate(relativeLummi = Luminescence / WT_Nullav)
WTFirefly_results_df
```

```r
WTFirefly_results_df <- WTFirefly_results_df %>% mutate(relativeLummi = Luminescence / WT_Nullav)
WTFirefly_results_df
E545KFirefly_results_df <- E545KFirefly_results_df %>% mutate(relativeLummi = Luminescence / E545K_Nullav)
E545KFirefly_results_df
H1047RFirefly_results_df <- H1047RFirefly_results_df %>% mutate(relativeLummi = Luminescence / H1047R_Nullav)
H1047RFirefly_results_df
```

```r
plotting_df <- rbind(WTFirefly_results_df, E545KFirefly_results_df, H1047RFirefly_results_df)
plotting_df
```

```r
p <- plotting_df %>% 
    ggplot(aes(x = Plasmid, y = relativeLummi)) +
    facet_wrap(~Genotype) +
    geom_bar(stat = "summary", position = "identity",
             aes(fill = Genotype, alpha = 0.6)) + 
    geom_point() +
#     coord_flip() + 
    scale_fill_manual(values = c("#C29ABB", "#B2D498", "#8CB3DF"),
                                     aes(alpha = 0.7)) +
#     geom_text(aes(y = 0.25, label = str_to_sentence(Desc)), size = 5, hjust = 0) +
    scale_y_continuous(expand = c(0,0)) + ylab(label = "Relative Luminescence") +
    theme_classic() + theme(text = element_text(family = "Arial"),
                          axis.title = element_text(size = 20),
                          axis.text = element_text(size = 16, angle = 45, hjust = 1,
                                                   colour = "black"),
#                           axis.text.y = element_blank(),
#                           axis.title.y = element_blank(),
#                           axis.ticks.y=element_blank(),
                          plot.title = element_text(size = 24),
                          plot.subtitle = element_text(size = 16, colour = "black"),
                          strip.text = element_text(size = 12),
                          legend.text = element_text(size = 14),
                          legend.title = element_text(size = 16),
                          legend.position = "none",
                          plot.margin = margin(10, 10, 10, 10))
p
```

```r
ggsave(filename="/data/park_lab/adam/figures/Luciferase9-8.png", 
       plot = p, type = "cairo", width = 8, height = 8)
```
