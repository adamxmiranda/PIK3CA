# anti-AREG Experiment

```r
library(tidyverse)
```

```r
setwd("/data/park_lab/adam/9.27.22 Anti-amp assay/")
```

```r
exp_data <- read_csv(file = "Summary__Sample Set (9-27-2022)_20220927_134650.csv")
exp_data
```

```r
exp_data_select <- exp_data %>%
                   dplyr::select(c("Sample ID", "Viable (x10^6) cells/mL"))
exp_data_select
```

```r
exp_data_select$Genotype <- c(rep("WT", 12), rep("E545K", 12), rep("H1047R", 12))
head(exp_data_select)
```

```r
exp_data_select$dosage <- c(rep("PBS", 3), rep("1ug", 3), rep("3ug", 3), rep("5ug", 3),
                            rep("PBS", 3), rep("1ug", 3), rep("3ug", 3), rep("5ug", 3),
                            rep("PBS", 3), rep("1ug", 3), rep("3ug", 3), rep("5ug", 3))
exp_data_select
```

```r
WT_DMSO_avg <- exp_data_select %>%
               dplyr::filter(Genotype == "WT") %>%
               dplyr::filter(dosage == "PBS")
E545K_DMSO_avg <- exp_data_select %>%
               dplyr::filter(Genotype == "E545K") %>%
               dplyr::filter(dosage == "PBS")
H1047R_DMSO_avg <- exp_data_select %>%
               dplyr::filter(Genotype == "H1047R") %>%
               dplyr::filter(dosage == "PBS")
WT_DMSO_avg <- mean(WT_DMSO_avg$`Viable (x10^6) cells/mL`)
E545K_DMSO_avg <- mean(E545K_DMSO_avg$`Viable (x10^6) cells/mL`)
H1047R_DMSO_avg <- mean(H1047R_DMSO_avg$`Viable (x10^6) cells/mL`)
WT_DMSO_avg
E545K_DMSO_avg
H1047R_DMSO_avg
```

```r
WTrelative <- exp_data_select %>%
               dplyr::filter(Genotype == "WT")
E545Krelative <- exp_data_select %>%
               dplyr::filter(Genotype == "E545K")
H1047Rrelative <- exp_data_select %>%
               dplyr::filter(Genotype == "H1047R")
```

```r
WTrelative$relativeSurvival <- WTrelative$`Viable (x10^6) cells/mL`/WT_DMSO_avg
WTrelative$relativeSurvival
E545Krelative$relativeSurvival <- E545Krelative$`Viable (x10^6) cells/mL`/E545K_DMSO_avg
E545Krelative$relativeSurvival
H1047Rrelative$relativeSurvival <- H1047Rrelative$`Viable (x10^6) cells/mL`/H1047R_DMSO_avg
H1047Rrelative$relativeSurvival
```

```r
WTrelative
E545Krelative
H1047Rrelative
```

```r
total_data <- rbind(WTrelative, E545Krelative, H1047Rrelative)
total_data
```

```r
library(ggsignif)
library(ggpubr)
library(ggprism)
```

```r
total_data$dosage <-factor(total_data$dosage, 
                                   levels = c("PBS", "1ug", "3ug", "5ug"))
```

```r
p <- total_data %>% #dplyr::filter(!Genotype =="WT") %>%
  ggplot(aes(fill = Genotype, x = dosage, 
                             y = relativeSurvival)) + 
  ylab("Relative Survival\nCompared to PBS Controls") +
  xlab("Concenctration of Antibody") +
  geom_bar(position = "dodge", stat = 'summary') + ylim(0, 1.4) +
  geom_point(aes(x = dosage), shape = 21, position = position_dodge(width = 1)) +
  geom_errorbar(stat = 'summary', position = position_dodge(width = .9), width = .25) + 
  stat_compare_means(aes(group=Genotype), method = "t.test", label.y = 1.3) +
#   geom_segment(aes(x = 0.75, xend = 1.25, y = 1.24, yend = 1.24)) +
  geom_segment(aes(x = 1.75, xend = 2.25, y = 1.1, yend = 1.1)) +
  geom_segment(aes(x = 2.75, xend = 3.25, y = 1.05, yend = 1.05)) +
  geom_segment(aes(x = 3.75, xend = 4.25, y = 0.75, yend = 0.75)) +
  geom_segment(aes(x = 2.75, xend = 2.75, y = 1.05, yend = 1.03)) +
  geom_segment(aes(x = 3.25, xend = 3.25, y = 1.05, yend = 1.03)) +
  geom_segment(aes(x = 2.25, xend = 2.25, y = 1.1, yend = 1.08)) +
  geom_segment(aes(x = 1.75, xend = 1.75, y = 1.1, yend = 1.08)) +
#   geom_segment(aes(x = 0.75, xend = 0.75, y = 1.24, yend = 1.22)) +
#   geom_segment(aes(x = 1.25, xend = 1.25, y = 1.24, yend = 1.22)) +
  geom_segment(aes(x = 3.75, xend = 3.75, y = 0.75, yend = 0.73)) +
  geom_segment(aes(x = 4.25, xend = 4.25, y = 0.75, yend = 0.73)) +
  ggtitle(label = "MCF10A Anti-AREG Treatment Screen", subtitle = "R&D Systems, MAB262-SP")
p
```

```r
total_data <- total_data %>% mutate(relativeSurvival = `Viable (x10^6) cells/mL` / 
                                    WT_control_avg)
total_data
```

```r
library(ggplot2)
library(ggpubr)
library(tidyverse)
library(ggpubr)
library(rstatix)
```

```r
stat.test <- total_data %>%
  group_by(Genotype) %>%
  t_test(relativeSurvival ~ dosage) %>%
  adjust_pvalue(method = "fdr") %>%
  add_significance("p.adj")
stat.test <- stat.test %>% mutate(xmin = rep(c(1, 1, 1, 2, 2, 3), 
                                             3)) %>%
             mutate(xmax = rep(c(2, 3, 4, 3, 4, 4), 
                                             3))
stat.test
```

```r
stat.test.sig <- stat.test %>% filter(!p.adj.signif == "ns")
stat.test.sig <- stat.test.sig %>% mutate(y.position = c(29, 26, 35, 33))
stat.test.sig
```

```r
EGF_exp_sel$Genotype <- c(rep("E545K", 12), rep("H1047R", 12), rep("WT", 12))
EGF_exp_sel$Media <- c(rep("not", 6), rep("maint", 6), 
                       rep("not", 6), rep("maint", 6),
                       rep("not", 6), rep("maint", 6))
EGF_exp_sel
```

```r
EGF_exp_maint <- EGF_exp_sel %>% filter(Media == "maint")
EGF_exp_maint
```

```r
WT_EGF_exp_maint <- EGF_exp_maint %>% filter(Genotype == "WT")
ex9_EGF_exp_maint <- EGF_exp_maint %>% filter(Genotype == "E545K")
ex20_EGF_exp_maint <- EGF_exp_maint %>% filter(Genotype == "H1047R")
WT_EGF_exp_maint
```

```r
WT_EGF_exp_maint <- WT_EGF_exp_maint %>% mutate(norm = .[[2]]/WT_PBS_av)
ex9_EGF_exp_maint <- ex9_EGF_exp_maint %>% mutate(norm = .[[2]]/ex9_PBS_av)
ex20_EGF_exp_maint <- ex20_EGF_exp_maint %>% mutate(norm = .[[2]]/ex20_PBS_av)
ex20_EGF_exp_maint
```
